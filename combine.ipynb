```python
import os
full_data=pd.DataFrame()
```

```python
for filename in os.listdir('processable data'):
    filepath=os.path.join('processable data', filename)
    df=pd.read_csv(filepath)
    full_data=pd.concat([full_data, df])
```

```python
full_data.head()
```

```python
len(full_data)
```

```python
full_data.to_csv('full_data.csv')
```
